```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
meat_to_animal
```

```python

data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# food값에 해당하는 데이터를 새로운 열에 추가
data['animal'] = data['food'].map(meat_to_animal) # 키를 기준으로 데이터 삽입
data
```

```python
def get_animal(x):
    return meat_to_animal[x]
# data['food'].map(get_animal)
data['animal'] = data['food'].map(get_animal)
# 위의 셀과 동일한 결과 
```

```python
# 연속형, 범주형
# 나이같은 경우: 연속형
# class로 나눠지는 데이터는 범주형

# 연속형 변수: 연속적인 값(나이, 점수, 몸무게, 가격, ...)
# 범주형 변수: 이산적인 값(서로 다른 것으로 구분이 되는가 = 성별, 혈액형, 학점(A, B+,..),...)
# 연속형 변수를 범주형 변수로 만들 수 있음(연령 24세 -> 20대)
```

```python
ages = [12, 13, 20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 37, 60, 100]
```

```python
age_category = pd.cut(ages, bins)
age_category
# ( ~ ] 개폐구간 = ( - 초과 , ] - 이하
```

```python
age_category.codes
```

```python
age_category.categories
```

```python
pd.cut(ages, bins, right=False) # 왼쪽을 포함하게 됨(range와 범위 표현이 같아짐)
```

```python
group_names = ['one', 'two', 'three', 'four']
pd.cut(ages, bins, labels= group_names)
```

```python
pd.describe_option() # 이 안에 있는 함수를 통해 출력 변경 가능
# pd.set_option('display.max_columns', 10)
```

```python
data = np.random.uniform(size=20)
data
```

```python
data_cut = pd.cut(data,4) # 동일한 4개의 범위를 분리
data_cut
```

```python
data_cut.codes
```

```python
pd.value_counts(data_cut.codes) # 구간별 카운트
```

```python
# qcut != cut
data_qcut = pd.qcut(data,4)
data_qcut
```

```python
pd.value_counts(data_qcut) # 구간별 동일하게 갯수 분리됨
```

```python
group_names = ['one', 'two', 'three', 'four']
data_cateq = pd.qcut(data,4,labels=group_names) # labels를 이용해서 
data_cateq
```
