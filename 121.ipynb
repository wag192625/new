```python
# 24.09.10
import pandas as pd
import numpy as np
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
# 피벗팅 pivot (자주 쓰임)
# 피봇테이블 : 데이터프레임 특정 열들을 행/열/데이터로 만드는 테이블
df1.pivot(index= '도시', columns= '연도', values= '인구')
```

```python
# set_index
df1.set_index(['도시','연도'])[['인구']].unstack()
```

```python
# df1.pivot(index='지역', columns='연도', values='인구') # index가 중복되는 데이터를 보유함으로 에러 출력
df1.pivot(index=['도시','지역'], columns='연도', values='인구')
# 위와 같은 멀티 인덱스를 갖도록 수정
# 행/열 인덱스는 데이터를 검색하기 위한 기준 => 값이 1개만 존재해야 함
```

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
# groupby
mygroup = df2.groupby('key1')
mygroup.groups # groups 쓰는 이유는 안쓰면 주소값만 출력력
```

```python
mygroup.sum() # 문자열은 합산이 안되서 미출력됨
```

```python
df2
```

```python
# df2의 key1을 기준으로 그룹화+data1의 합계
df2.groupby('key1')['data1'].sum()

df2.data1.groupby(df2.key1).sum() # 위와 동일한 출력
# df2.data1.groupby('key1').sum() # data1한테 key1이 없고 data1의 열을 이미 사용했기 때문에 오류

# df2.groupby('key1').sum().data1 # 권장하지 않는 방법
```

```python
df2.groupby(['key1','key2'])['data1'].sum()

# 위와 같은 출력
# df2.groupby(['key1','key2']).data1.sum()

df2.data1.groupby([df2.key1, df2.key2]).sum()
```

```python
df2.groupby(['key1', 'key2']).sum().data1
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.groupby('species').sum()
```

```python
iris.groupby('species').mean() # 그룹별 각 열에 대한 평균값
```

```python
iris.groupby('species').median() # 중위값
```

```python
iris.groupby('species').min()
```

```python
iris.groupby('species').max()
```

```python
iris.groupby('species').std() # 표준편차
```

```python
iris.groupby(iris.species).var() # 분산
```

```python
iris.groupby('species').quantile() # 4분위수, 중위수와 동일하게 출력
```

```python
iris.groupby(iris.species).median() # quantile과 동일함
```

```python
iris.groupby('species').first() # 그룹의 첫번째 데이터를 이용
iris.groupby('species').last() # 그룹의 마지막 데이터를 이용
# 잘 안쓰임
```

```python
iris.groupby('species').groups # 종류별로 데이터 50개
```

```python
iris.groupby('species').count() # 셀 별로 데이터 건수 확인
```

```python
iris.groupby('species').size() # 그룹별 데이터 갯수

# count, size 차이
# - count는 열별로, size는 그룹별 개수 출력
# - size는 NaN을 포함
```

```python
df1.pivot_table('인구','도시','연도',margins=True) # 집계함수 진행 -> 평균 구하기
# 3.451492e+06 => 3.451492 10**6
# pd.options.display.float_format = '{:.2f}'.fortmat 하면 소수점 2자리까지만 출력
```

```python
df1.pivot_table('인구','도시','연도',margins=True, aggfunc='max', fill_value='1') # 최대값 추출
# aggfunc과 fill_value는 직접 다른 함수를 선언하여 입력하거나 다른 함수를 입력해도 무관
```
