1. 배열(Array)
    기본적인 데이터 구조를 가지고 있음. 보통 사이즈를 정해두고 만듦. 범위 내의 효율은 극강
    인덱스 값을 가져오는 방식은
    big O 방식으로 설명함.
    가져오는 방법
    가져오기, 추가, 삭제
    시간복잡도 : O(n)

    배열 특징
    장점
    - 만들고 사용하기 쉽다.(int arr = [])
    - 트리나 정렬, 리스트 관련한 다양한 구조 및 알고리즘의 기초가 되는 구조
    - 원하느 데이터를 빠르게 효율적으로 조회 및 사용 가능
    - 정렬을 하는데 굉장히 용이함

    단점
    - 데이터를 저장하는데 크기가 고정되어 있음
    - 크기를 변경할 때 효율이 떨어짐(비효율적)
    - 데이터를 삭제하거나 추가하는 방법은 다소 비효율적임

    테이블, 벡터, 행렬 이런 구조를 구현하는데 사용(이외에도 다양한 구조에 사용)

2. 스택(Stack = 파이썬의 pop() = LIFO Last In First Out)
    LIFO 방식의 사용법을 적용.
    가장 최근에 들어온 데이터를 처리(접시)
    stack의 시간 복잡도
    가져오기, 추가, 삭제
    O(n)     O(1)    O(1)

    스택 특징
    장점
    - 동적인 메모리 크기
    - 데이터가 들어온대로 순서가 정렬(list.append와 같음)
    - 빠른 런타임(실행이 빠름)

    단점
    - 최신 데이터만 가져올 수 있음(이게 n으로 만드는 원인)
    - 한번에 단 하나의 데이터만 처리 가능

    브라우저의 뒤로가기, 실행 취소, 재귀 함수, 마지막 데이터부터 순차적 처리할 때

3. 큐(Queue)
    스택과 반대로 진행한다.(FIFO, 선입선출)
    가장 먼저 들어온 데이터를 처리하는 매커니즘을 가짐.
    놀이공원, 매장, 식당 등 순서대로 입장하고 처리하는 경우에 사용되는 방식.
    O(n), O(1), O(1)
    
    장점
    - 동적인 메모리 크기
    - 데이터를 받는 순서대로 정렬
    - 빠른 런타임

    단점
    - 가장 오래된 요소를 가져옴 (큐와 스택의 차이점)
    - 한번에 하나만 처리 가능

    쓰임처
    - 반복적이고 자주 받는 데이터를 처리할 때 효율적
    - 순서에 민감한 데이터를 처리할 때 중요(텍스트, 음성(자연어),... 등)
    - 캐시(cache) 구현

4. 연결리스트(Linked List) = 참조를 하는 구조로 되어있음
    node가 존재하며 여기애 저장함(주소나 포인터를 저장)
    모든 노드가 연결되는(연결되도록 반복) 구조
    이에 따라서 데이터의 추가나 삭제 시 재구성을 할 필요가 없음
    종류가 나뉘는데 단일 연결 리스트, 이중 연결 리스트, ...

    시간복잡도
    가져오기, 삭제, 추가
    O(n)    O(1)    O(1)

    장점
    - 새로운 데이터의 추가/삭제가 효율적임
    - 배열처럼 메모리에 연속적인 위치를 가지진 않는다.
    - 배열과 같은 구조의 재구성이 필요 없음
    - 동적인 메모리(메모리가 한군데에 묶여있을 필요가 없음)
    - 메모리를 효울적으로 사용 가능 -> 빅데이터 처리 적합
    
    단점
    - 배열보다 메모리를 더 사용함(집에서 출가한 자식들이 집 한채씩 가지고있어야 하니 집이 늘어남)
    - 처음부터 끝까지 순회. 이에 따라 지목한 데이터를 탐색하는데에는 효율이 낮음
        메모리가 붙어있지 않기 때문에 하나씩 다 찾아야함 (O(n))
    - 노드를 반대 방향으로 검색할 때 비효율적임(이중 연결리스트일 경우)
        [[][]],[[1][][]], [[]]
        1을 찾고 싶은데 ['[]']를 먼저 찾고 '[[]]' 검색 후 [[][]'[]'] 검색 반복
    
    사용처
    - 메모리 크기가 정해지지 않았을 때
    - 데이터를 연속적으로 빠르게 삽입/제거할 때
    - 이미지 뷰어, 갤러리, 음악 등 탐색

5. 해시 테이블(딕셔너리)
    대량의 정보를 저장하고 특정 요소를 효율적인 탐색을 할 수 있는 복잡하게 설계된 데이터 구조(키-값 : 쌍)
    키는 검색 시  사용되고, 값은 키의 해당하는 데이터를 의미함.

    시간복잡도
    가져오기, 삭제, 추가
    O(1), O(1), O(1)

    장점
    - 새로운 정보를 추가하거나 삭제에 용이하고 효율적임
    - 검색, 가져오기가 빠름
    - 동적인 메모리(크기는 직접 조절해야됨)

    단점
    - 충돌이 (자주)일어날 수 있음(키 값이 동일하면 안됨)

    사용처
    - 주소, 사전, 번호, 로그인, 기타 등등

6. 그래프
    각 node 사이에 edge가 있는 구조 (0-0-0   0이 노드 - 엣지)
    방향이 있는 경우(directed)와 없는 경우가 있음(undirected)
    방향이 있는건 단방향(일방통행)
    방향이 없는 경우양방향(쌍방통행)

    <!-- 지도(최적 경로 탐색) -->

    시간복잡도가 매번 다름(구조 설계나 목표가 무엇인가에 따라)
    시간복잡도
    가져오기        O([node] + [edge]) 노드의 갯수와 엣지의 갯수만큼 시간이 걸림
    노드/엣지 추가  O(1)
    노드 삭제       O([node] + [edge])
    엣지 삭제       O([E])

    장점
    - 새로운 요소의 추가/삭제가 용이하고 효율적
    - 구조의 응용 가능
    
    - 충돌이 일어날 수 있음(키 값의 영향을 받음)

    사용처
    - DB : 주소찾기, 이름, 전화번호, 기타 등등
    - 사용자 로그인 인증,...

    연결, 해시 테이블보다 어려운 상황에서 사용

7. 트리
    노드로 구성된 계층적 자료구조.(컴퓨터의 디렉토리(폴더))

    개념들
    - 트리의 구성요소를 node라고 한다.
    - 노드와 노드를 잇는 선은 edge라고 함(그래프와 같음)
    - 자식이 없는 노드는 leaf

          0     depth(깊이) 0 (최상위 노드는 root라고 부름)
         / \    
        0   0   depth 1
       ... ...  depth 2 (silblings = 같은 부모를 가진 노드들)
    
    자식이 2개 이하면 이진 트리(완전히 이진트리로 구성되면 완전 이진 트리)
