```python
# 포맷팅(format)
import datetime
month = 5
day = 7 #날짜 설정 된 상태
print('오늘의 날짜는 {}월 {}일'.format(month,day)) # 포맷팅 1
print('오늘의 날짜는 {day}월 {month}일'.format(month = 5 ,day= 7)) # 포맷팅 2
print(f'오늘의 날짜는 {month}월 {day}일')

name = '정명훈'
age = 26
print(f'나의 이름은 {name}이고 {age}이다')
print('나의 이름은 {}이고 {}이다'.format(name, age))
print('나의 이름은 {name}이고 {age}이다'.format(name = '정명훈', age = 26))
print('나의 이름은 {1}이고 {0}이다'.format(4,6))
```
