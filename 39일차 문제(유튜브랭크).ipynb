https://youtube-rank.com/board/bbs/board.php?bo_table=youtube

위 주소에서 유튜버를 구독자수로 정렬했을 때 
순위, 채널명, 구독자수 를 추출

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
wd = webdriver.Chrome() # 크롬창 열기 (wd는 웹드라이버 약칭)
```

```python
url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube'
wd.get(url) # 크롬창에서 유튜브 랭킹
```

```python
html = wd.page_source   # 랭킹 차트의 소스를 html로 지정
soup = BeautifulSoup(html, 'html.parser')   # 뷰티풀수프로 파싱
```

```python
soup.select('#list-skin > form:nth-child(5) > table > tbody > tr h1 a')[0].text.strip()    #채널명
soup.select('#list-skin > form:nth-child(5) > table > tbody > tr > td.subscriber_cnt')[0].text # 구독자 수
```

```python
ranking_data = []
for i in range(50):
    rank = i+1
    channel = soup.select('#list-skin > form:nth-child(5) > table > tbody > tr h1 a')[i].text.strip()
    subscriber = soup.select('#list-skin > form:nth-child(5) > table > tbody > tr > td.subscriber_cnt')[i].text
    ranking_data.append([rank,channel,subscriber])
ranking_data
```

```python
import pandas as pd # 판다스

df = pd.DataFrame(ranking_data, columns=['순위', '채널', '구독자 수'])
df.tail()
df.to_excel('youtube rank50.xlsx', index=False) # 액셀로 추출 /인덱스는 왼쪽 숫자 제거
```
