```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
wd = webdriver.Chrome()
```

```python
url = 'https://store.steampowered.com/charts/topselling/KR'
wd.get(url)
```

```python
html = wd.page_source   # 랭킹 차트의 소스를 html로 지정
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('#page_root > div:nth-child(5) > div > div > div > div._3sJkwsBQuiAc_i3VOWX4tv > table > tbody tr td a')[0].text
soup.select('#page_root > div:nth-child(5) > div > div > div > div._3sJkwsBQuiAc_i3VOWX4tv > table > tbody > tr > td._3IyfUchPbsYMEaGjJU3GOP > div > div ')[0].text
```

```python
game_ranking = []
for i in range(50):
    rank = i+1
    title = soup.select('#page_root > div:nth-child(5) > div > div > div > div._3sJkwsBQuiAc_i3VOWX4tv > table > tbody tr td a')[i].text
    price = soup.select('#page_root > div:nth-child(5) > div > div > div > div._3sJkwsBQuiAc_i3VOWX4tv > table > tbody > tr > td._3IyfUchPbsYMEaGjJU3GOP > div > div')[i].text
    game_ranking.append([rank, title, price])
game_ranking
```

```python
import pandas as pd # 판다스

df = pd.DataFrame(game_ranking, columns=['순위', '타이틀', '가격'])
df
```
