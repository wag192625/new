```python
# 38일차 다시 풀기
# 3.
# 1차원의 점들이 주어졌을 때, 그 중 가장 거리가 짧은 것의 쌍을 출력하는 함수를 작성하시오. (단 점들의 배열은 모두 정렬되어있다고 가정한다.)
# 예를들어 S={1, 3, 4, 8, 13, 17, 20} 이 주어졌다면, 결과값은 (3, 4)가 될 것이다.
# S={1, 3, 4, 8, 13, 17, 20}
s=[1, 3, 4, 8, 13, 17, 20]
lst=[]
num = 0
aaa = 0
bbb = []
for i in s :
    if i != s[0]:
        num+=1
        aaa = i-s[num-1]    # 서로 차이
        # print('i=', i,'s=',s[num-1])
        # print('aaa=',aaa)
        bbb.append(aaa)
        print(bbb)
        if aaa < bbb[0] :
            lst.append(i)
            lst.append(s[num-1])
print(lst)
```
