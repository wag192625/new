```python
df = pd.read_csv('data/titanic/train.csv')
df
```

```python
df.info()
df.describe()
```

```python
df = df.drop('PassengerId', axis=1)
df
```

```python
def get_nickname(x):
    nickname = x.split('.')[0]
    nickname = nickname.split()[-1]
    return nickname

df['title'] = df.Name.apply(get_nickname)
df
```

```python
df.drop('Name', axis=1, inplace=True)
df
```

```python
df.title.value_counts().index
```

```python
title_mapping = {'Mr':1, 'Miss':2, 'Mrs':3, 'Master':4, 'Dr':0, 'Rev':0, 'Mlle':0, 'Major':0, 'Col':0,
       'Countess':0, 'Capt':0, 'Ms':0, 'Sir':0, 'Lady':0, 'Mme':0, 'Don':0, 'Jonkheer':0}
```

```python
df['title'] = df['title'].map(title_mapping)
df
```

```python
# 'Mr':1, 'Miss':2, 'Mrs':3, 'Master':4,
df[['title','Survived']].value_counts().sort_index()
```

```python
df.drop('Ticket', axis=1, inplace=True)
df
```

```python
df.drop('Cabin', axis=1, inplace=True)
df
```

```python
df['family_size'] = df.SibSp + df.Parch + 1
df
```

```python
sex_mapping = {'male': 0, "female": 1}
df['Sex'] = df['Sex'].map(sex_mapping)
df
```

```python
df[['family_size','Survived']].value_counts().sort_index()
```

```python
df.Embarked.value_counts()
```

```python
df['Embarked'] = df.Embarked.fillna('S')
df.info()
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
df
```

```python
embark_mapping = {'S': 0, "C": 1, "Q": 2}
df['Embarked'] = df.Embarked.map(embark_mapping)
df
```

```python
# scikit-learn: 머신러닝 라이브러리(CPU로 동작)
# tensorflow, pytorch : 딥러닝 라이브러리
```

```python
from sklearn.neighbors import KNeighborsClassifier # 종류 예측 모델(분류모델) != 선형 회귀(회귀 모델)
clf = KNeighborsClassifier(n_neighbors=7) # 인접한 데이터의 참고 개수
```

```python
# 교차검증
from sklearn.model_selection import KFold, cross_val_score
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=2024) # 교차검증 10회 진행
```

```python
train = df.drop('Survived', axis=1)
train
```

```python
target = df.Survived
target
```

```python
score = cross_val_score(clf, train, target, cv=k_fold, n_jobs=1, scoring='accuracy')
print(score)
```

```python
score.mean()
```
