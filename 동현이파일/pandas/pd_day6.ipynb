```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# 데이터 합치기
pd.merge(df1, df2) # 공통적 열을 기준으로 합함. 겹치지 않는 열은 제외(inner join)
```

```python
pd.merge(df1, df2, how='outer') # outer join
```

```python
pd.merge(df1, df2, how='left') # 왼쪽 파라미터를 기준으로 병합
```

```python
pd.merge(df1, df2, how='right') # df2 기준으로 병합
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2, on='고객명') # 많이 쓰임
# 중복되는 열이 2개 이상일 때 겹치는 데이터에 대해서 탐색시도하므로 출력할 데이터가 없게 됨
# 파라미터에 추가로 on=컬럼명 진행하면 하나의 컬럼만 중심으로 합칠 수 있음
```

```python
merged_df = pd.merge(df1, df2, on='고객명')
merged_df.rename(columns={'데이터_x': '데이터', '데이터_y' : '성별'}, inplace=True)
merged_df
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on='성명')
# 같은 열이름을 가진 데이터가 없어서 원래는 merge가 안됨
# left_on과 right_on으로 서로 다른 열이지만 같은 열처럼 명시해주면 merge할 수 있음
```

```python
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['데이터1', '데이터2'])
df2
```

```python
pd.merge(df1, df2, left_on=['도시','연도'], right_index=True) # 흔치 않음
```

```python
pd.merge(df2, df1, right_on=['도시','연도'], left_index=True)
```
