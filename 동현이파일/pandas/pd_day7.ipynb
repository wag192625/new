```python
import pandas as pd
import numpy as np
```

```python
#1. df1과 df2에 대해 merge 연습, on, left_on, right_on, how
#2. left과 right에 대해 merge 연습, on, left_on, right_on, how
#3. merge 열 이름과 행 인덱스를 기준으로 병합 연습
#4. merge연습, index 기준
#5. concat 연습, axis, join 속성 값 변경 설정
#6. stack, unstack 연습
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['데이터3', '데이터4'])
df1, df2
```

```python
# pd.merge(df1, df2) # 중복되는 열(column)이 없어서 안됨
pd.merge(df1, df2, left_index=True, right_index=True) # 중복되는 행(index) 기준 병합
pd.merge(df1, df2, left_index=True, right_index=True, how='outer') # 병합 데이터 없을 시 NaN값 출력

pd.merge(df1, df2, left_on='데이터1', right_on='데이터3',how='outer') # 열 기준 병합(a,b,c 행 사라짐)
pd.concat([df1,df2])
pd.concat([df1,df2], axis=1)
```

```python
# pd.merge(df1, df2) # 중복되는 열(column)이 없어서 안됨
pd.merge(df1, df2, left_index=True, right_index=True) # 중복되는 행(index) 기준 병합
pd.merge(df1, df2, left_index=True, right_index=True, how='outer') # 병합 데이터 없을 시 NaN값 출력

pd.merge(df1, df2, left_on='데이터1', right_on='데이터3',how='outer') # 열 기준 병합(a,b,c 행 사라짐)
pd.concat([df1,df2])
pd.concat([df1,df2], axis=1)
```

```python
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['데이터1', '데이터2'])
df1
# df2
```

```python
df1.stack() # 도시, 연도, 인구 열이 하나의 행에 하위 행으로
```

```python
df1.unstack() # 도시, 연도, 인구가 각각의 행으로
```

```python
df11 = df1.set_index('도시')
pd.merge(df11,df2, left_on=['도시','연도'], right_index=True)
pd.merge(df11,df2, left_on=['도시','연도'], right_index=True, how='outer')
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
# 피벗팅 pivot(자주 쓰임)
# 피봇테이블: 데이터프레임 특정 열들을 행/열/데이터로 만드는 테이블
df1.pivot(index='도시', columns='연도', values='인구') # 최신 버전에서 돌아가는 코드
```

```python
# set_index
df1.set_index(['도시','연도'])[['인구']].unstack()
```

```python
# df1.pivot(index=['지역'], columns='연도', values='인구') # index가 중복되는 데이터를 보유하므로 에러 출력
df1.pivot(index=['도시','지역'], columns='연도', values='인구')
# 위와 같은 멀티인덱스를 갖도록 수정
# 행/열 인덱스는 데이터를 검색하기 위한 기준 => 값이 1개만 존재해야 함
```
