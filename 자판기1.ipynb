```python
# 자판기 프로그램(사용자 모드, 관리자 모드, 재고 관리 기능 등)
#  사용자모드 => 돈 넣고 돈 넣은만큼 음료 고르기, 잔돈 교환
#  관리자모드 => 제품 추가/제거, 가격조정
#  재고 관리 기능 => 제품 재고 확인 및 추가/제거
drink = ['콜라', '사이다', '환타', '몬스터', '핫식스', '비타500']
price = [1800, 1600, 1600, 2200, 2100, 1900]
stock = [7,10,13,7,10,5]
drink_price = dict(zip(drink,price))
drink_stock = dict(zip(drink,stock))

# 사용자
def vending_machine():
    money = 4000
    # money = int(input())
    # # def vending_machine():
    while(money>=0):
        print(drink_price)
        choice_drink = input()
        if drink_stock.get(choice_drink) == 0 :
            print('품절입니다.')
        elif drink_price.get(choice_drink) > money :   #돈이 제품의 가격보다 낮으면 
            print('잔액이 부족합니다. 현재 잔액은', money, '입니다.' )
        elif choice_drink == 'end' :
            
            print('거스름돈',money,'반환합니다.')
            break
        elif choice_drink in drink_price:
            money-=drink_price.get(choice_drink)
            drink_stock[choice_drink] = drink_stock.get(choice_drink) -1
            print(choice_drink,'선택하셨습니다. 잔액은',money,'원입니다.')

# 관리자
def manager():
    while(True):
        print(drink_price)
        print(drink_stock.values)
        input_word = input()
        if input_word == 'price':   # 가격 조정         애초에 제품명 입력하고 그 안에서 뭔가 바꾸는게 더 좋을거 같음
            item_name = input()     # 조정할 상품 이름
            print(item_name)
            print(drink_price.get(item_name))
            item_price = int(input())   #조정할 상품 가격
            drink_price[item_name] = item_price

        elif input_word == 'add' :  # 제품 추가
            print('제품명 입력')
            add_item_name = input() #추가할 상품 이름   이것도 위와 같음
            print('가격 입력')
            add_item_price = int(input())   #추가할 상품 가격
            drink_price[add_item_name] = add_item_price 
            drink_stock[add_item_name] = 0  #재고는 0으로 
            print(drink_price,drink_stock)

        elif input_word == 'del' :  #상품 제거
            print('제품명 입력')
            del_item_name = input() # 제거할 상품 이름
            del drink_price[del_item_name]
            del drink_stock[del_item_name]
            print()
        else :
            break

# # 재고 관리
def inventory_management() :
    while(True) :
        print(drink_stock)
        input_word = input()    # 상품 이름
        if input_word == 'end':
            break
        input_stock = int(input)    # 상품 가격 조정
        drink_stock[input_word] = input_stock       # 한꺼번에 조정하면 좋을거 같음
    
aaa= input()
# while 문으로 돌려서 계속 돌아가게, 기본은 사용자 모드고 사용자 모드에서 관리자, 재고 진입하게끔 만들
if aaa == '사용자' :    # 엔터 치면 바로 자판기 이용하게 하려 했는데 None 했는데 안됨
    vending_machine()
elif aaa == '관리자' :
    manager()
elif aaa == '재고관리':
    inventory_management()
```
